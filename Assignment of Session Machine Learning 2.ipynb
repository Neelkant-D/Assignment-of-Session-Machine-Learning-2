```python
bos['PRICE'] = boston.target
bos.head()
```

```python
from sklearn.linear_model import LinearRegression

x = bos.drop('PRICE', axis =1 )  ## x = All Features,except price
y=  bos.PRICE                   ## y is housing price
lm =LinearRegression()
lm.fit(x,y)

```

```python
# print intercept and coefficients

print('Estimated intercept coeffiecient',lm.intercept_)
print('Number of Coeffiecients',len(lm.coef_))
```

```python
z= lm.predict(x) ## Predicted values stored in container z
z
```
